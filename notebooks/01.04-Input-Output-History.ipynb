# Data overview

Something about the data here

```python
import requests
import pandas as pd

api_url = "https://charts.ecmwf.int/opencharts-api/v1/"
result =requests.get('{}products/'.format(api_url))
products = pd.json_normalize(result.json())
```

```python
from IPython.display import display, HTML

def display_with_scroll(df, height=300, header_bg_color='#4CAF50', header_text_color='white'):
    display(HTML(f"""
    <div style="height:{height}px; overflow-y:scroll;">
        <style>
            table {{text-align: left; border-collapse: collapse; width: 100%;}}
            th, td {{text-align: left !important; padding: 8px; border: 1px solid #ddd;}}
            th {{
                background-color: {header_bg_color};
                color: {header_text_color};
            }}
            tr:nth-child(even) {{background-color: #f2f2f2;}}
        </style>
        {df.to_html(index=False)}
    </div>
    """))

display_with_scroll(products, height=300)
```
